## Основные теоретические положения

### Нормальный случайный вектор
**Плотность многомерного нормального случайного вектора**: 

$ p_{\vec{\xi}} (\vec{x}) = \cfrac{1}{ (2\pi)^{n/2} \cdot \sqrt{\det \Sigma}} \cdot exp \left( - \cfrac{1}{2} \cdot (\vec{x} - \mu)^{\intercal} \cdot \Sigma^{-1} \cdot (\vec{x} - \mu) \right) \quad \quad (1) $


В заданной плотности нормального вектора **константу** $\mathbf{С}$, если она неизвестна, соответственно можно найти из формулы $(1)$:

$ C = \cfrac{1}{ {(2\pi)}^{ {n}/{2} } \cdot \sqrt{\det \Sigma} } $

Одним из **свойств многомерного нормального случайного вектора**, которое будет использоваться в решении задач, является:

$ \vec{\xi} \sim \mathscr{N} (\mu, \Sigma), \vec{\xi} = (\xi_1, \dotso, \xi_n)^{\intercal}, A=(a_{ij})_{1 \leq i,j \leq n}, b=(b_1, \dotso, b_n)^T : \vec{\eta} = A \vec{\xi} + b \sim \mathscr{N} (A \cdot \mu + b, A \Sigma A^T) $

### Вектор математических ожиданий и ковариационные характеристики нормального случайного вектора

В задании данной практической работы в выражении плотности двумерного нормального случайного вектора значение степени экспоненты задано не в явном виде, как $ - \cfrac{1}{2} (\vec{x} - \mu)^{\intercal} \cdot \Sigma^{-1} \cdot (\vec{x} - \mu)$ в формуле $(1)$, однако **исходное выражение можно привести к виду**: 

$ -\cfrac{1}{2} (\alpha (x-a)^2 + \beta (x-a)(y-b) + \gamma (y-b)^2) \quad \quad (2) $

Из $(2)$ выражения мы можем получить значения для **вектора математических ожиданий** и **матрицы** $ \mathbf{\Sigma^{-1}} $:

$ \quad 
\mathbb \mu = 
    \begin{pmatrix}
        a \\
        b
    \end{pmatrix}
\quad
\Sigma^{-1} = 
    \begin{pmatrix}
        \alpha & \beta / 2\\
        \beta / 2 & \gamma
    \end{pmatrix}
$


**Матрицу ковариации** $ \mathbf{\Sigma} $ можно найти из матрицы $ \Sigma^{-1} $ следующим образом:

$ \Sigma = \cfrac{1}{\det \Sigma^{-1}} 
 \begin{pmatrix}
    \gamma & - \beta/2 \\
    - \beta/2 & \alpha
\end{pmatrix} $


На главной диагонали такой матрицы находятся **значения дисперсий $ \mathbf{\xi} $ и $ \mathbf{\eta} $**, а вне её - **значения ковариации**:

$\Sigma = 
 \begin{pmatrix}
  \mathbb D \xi & \mathop{\mathrm{cov}}\nolimits(\xi, \eta)  \\
  \mathop{\mathrm{cov}}\nolimits(\xi, \eta)  & \mathbb D \eta \\ 
 \end{pmatrix}$

**Коэффициент корреляции**: 

$ \rho(\xi, \eta) = \cfrac{\mathop{\mathrm{cov}}\nolimits(\xi, \eta) }{\sqrt{\mathbb D \xi \cdot \mathbb D \eta}} $

### Стандартный нормальный случайный вектор

**Стандартный нормальный случайный вектор** имеет вид: $ \quad \vec{\xi} \sim \mathscr{N} (0, 1) $

Таким образом, **вектор математических ожиданий и матрица ковариаций**:

$
\mu_{st} = 
    \begin{pmatrix}
        {0} \\
        {0}
    \end{pmatrix}
\quad
\Sigma_{st} = 
 \begin{pmatrix}
  1 & 0 \\
  0 & 1 \\ 
 \end{pmatrix}$

Чтобы получить стандартный нормальный случайный вектор над исходным случайным вектором необходимо провести **аффинное преобразование**, то есть найти такую ненулевую матрицу $ A $ и вектор $ b $, что 

$ \mu_{st} = A \mu + b \quad \text{и} \quad \Sigma_{st} = A \Sigma A^{\intercal}  $ 


### Необходимые формулы для ортогональных преобразований

$ \det(\Sigma^{-1} - \lambda \cdot E) = 0 \quad \quad (3)$

Из выражения выше находим **собственные числа** $\lambda_1$ и $\lambda_2$, после подставляем их в исходную матрицу $\Sigma^{-1}$, из которой составляем систему уравнений, и находим собственные вектора, не равные вектору $(0, 0)^{\intercal}$. 

**Собственные вектора** необходимо нормировать по формуле:

$ \vec{v} = \alpha 
\begin{pmatrix}
    a\\
    b
\end{pmatrix} = \cfrac{1}{\sqrt{a^2+b^2}} \cdot 
\begin{pmatrix}
    a\\
    b
\end{pmatrix}
$

Из собственных векторов составляется матрица $Q^{\intercal}$.

В данной работе необходимо выполнить **ортогональное преобразование вида**:

$ \Sigma_{id} = Q \Sigma Q^{\intercal} \quad \quad (4)$ 

На диагонали полученной матрицы $ \Sigma_{id} $ будут стоять значения, обратные по умножению значениям собственных векторов $\lambda_1$ и $\lambda_2$.

### Условное распределение
**Плотность условного распределения** для непрерывных случайных векторов находится по формуле:

$ p_{\xi | \eta=y_0} (x) = \cfrac{ p_{\xi, \eta}(x, y_0) }{ p_\eta  (y_0) } \quad \quad (5)$

**Нормальное одномерное распределение** для компоненты $\eta$ будет иметь вид:

$ \eta \sim \mathscr{N}(a, \sigma^2) $

$ p_\eta (y) = \cfrac{1}{\sigma\sqrt{2\pi}} \cdot \exp \left(-\cfrac{1}{2}\left(\cfrac{y-a}{\sigma}\right)^2\right) \quad \quad (6)$

***Примечание***: для нормального распределения $ \mathbb{E} \eta = a, \mathbb{D} \eta = \sigma^2$

Подставим полученные значения в формулу $(5)$ из теории выше:

$p_{\xi|\eta=y_0} (x) = \cfrac{ \cfrac{3}{\pi} \cdot \exp \left(-\cfrac{1}{2}(5x^2-4xy_0+8y_0^2-6x-12y_0+9)\right) }{ \cfrac{6}{\sqrt{10 \pi}} \cdot \exp \left(-\cfrac{1}{2}\left(\cfrac{36}{5}y_0^2 - \cfrac{72}{5}y_0 + \cfrac{36}{5}\right)\right) } = \cfrac{\sqrt{10 \pi}}{2 \pi}  \cdot \exp \left(-\cfrac{1}{2} \left( 5x^2-4xy_0 + \cfrac{4}{5}y_0^2 -6x + \cfrac{12}{5}y_0 + \cfrac{9}{5} \right)\right) = \cfrac{1}{ \cfrac{1}{ \sqrt{5} } \cdot \sqrt{2 \pi}} \cdot \exp \left(-\cfrac{1}{2} \left(\cfrac{ x-\cfrac{4y_0+6}{5} }{ \cfrac{1}{ \sqrt{5} } } \right)^2\right)$

Используя примечание к формуле $(6)$ из теории выше, получаем, что условные математическое ожидание и дисперсия равны:

$ \mathbb{E} (\xi|\eta) = \cfrac{4y_0 + 6}{5} $

$ \mathbb{D} (\xi|\eta) = \cfrac{1}{5} $
